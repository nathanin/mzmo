## Histology image data:

I have two folders: one with all tiles from M0 and one with all tiles from M1:

* M0 has 1200 tiles  
* M1 has 2046 tiles 

Run all through AlexNet with transferred weights from the ImageNet classificaiton task

```python
m0_path = '/Users/nathaning/local_projects/mzmo/data/m0_256'
m1_path = '/Users/nathaning/local_projects/mzmo/data/m1_256'

m0_images = glob.glob(os.path.join(m0_path, '*.tif'))
m1_images = glob.glob(os.path.join(m1_path, '*.tif'))
```

Transfer learning network fine-tuned to the M0/M1 task

```python
m0_correct = 0
m1_correct = 0


for m0_img in m0_images:
    img = get_img(m0_img, 227) # Resize each 256 --> 227
    net_transfer, label = infer(net_transfer, img)
    m0_correct+=(label == 0) 
    #print "Image {} classified as \t{}".format(m0_img, label)
    
for m1_img in m1_images:
    img = get_img(m1_img, 227) # Resize each 256 --> 227
    net_transfer, label = infer(net_transfer, img)
    m1_correct+=(label == 1) 
    #print "Image {} classified as \t{}".format(m1_img, label)
    
    
print "Number of M0 images classified as M0: {}".format(m0_correct)
print "Number of M1 images classified as M1: {}".format(m1_correct)
```

End to end training on the M0 / M1 task

```python
m0_correct = 0
m1_correct = 0


for m0_img in m0_images:
    img = get_img(m0_img, 227) # Resize each 256 --> 227
    net_end2end, label = infer(net_end2end, img)
    m0_correct+=(label == 0)
    #print "Image {} classified as \t{}".format(m0_img, label)
    
for m1_img in m1_images:
    img = get_img(m1_img, 227) # Resize each 256 --> 227
    net_end2end, label = infer(net_end2end, img)
    m1_correct+=(label == 1)
    #print "Image {} classified as \t{}".format(m1_img, label)
    
    
print "Number of M0 images classified as M0: {}".format(m0_correct)
print "Number of M1 images classified as M1: {}".format(m1_correct)
```

GoogLeNet, trained end to end on the M0/M1 task

```python
m0_correct = 0
m1_correct = 0

for m0_img in m0_images:
    img = get_img(m0_img, 224) # Resize each 256 --> 224
    net_google, label = infer(net_google, img, 'softmax')
    m0_correct+=(label == 0) 
    #print "Image {} classified as \t{}".format(m0_img, label)
    
for m1_img in m1_images:
    img = get_img(m1_img, 224) # Resize each 256 --> 224
    net_google, label = infer(net_google, img, 'softmax')
    m1_correct+=(label == 1) 
    #print "Image {} classified as \t{}".format(m1_img, label)
    
    
print "Number of M0 images classified as M0: {}".format(m0_correct)
print "Number of M1 images classified as M1: {}".format(m1_correct)
```

## Visualize the filters and net activations

Net filters are a visualization of the weights within each convolution layer. The input image is passed through the filter, and the elements of the image which respond to it are allowed to pass through, so to speak. Once training is done, the filters are set. 

Filters from ImageNet trained AlexNet

```python
# the parameters are a list of [weights, biases]
filters = net_bvlc.params['conv1'][0].data
vis_square(filters.transpose(0, 2, 3, 1))

# The parameters are fixed once training is finished
```

Activation of each filter in response to the input image. Brighter areas mean it responded strongly. In the first layer, we can still see large objects obviously. Going to deeper convolutions, more of the original image impacts each pixel in the activation. Therefore, it quickly becomes difficult or impossible to recognize parts of the original image in deeper convolutions. 

ImageNet activation (Convolution 1) to the above image:

```python
# Using the BVLC-AlexNet 
feat = net_bvlc.blobs['conv1'].data[0, :]
vis_square(feat)
```

Filters for ImageNet transferred weights, allowed to fine-tune on the M0 / M1 task:

```python
# the parameters are a list of [weights, biases]
filters = net_transfer.params['conv1'][0].data
vis_square(filters.transpose(0, 2, 3, 1))

# The parameters are fixed once training is finished
```

And the first convolution activations:

```python
# Using the Transferred / Fine tuned model
feat = net_transfer.blobs['conv1'].data[0, :]
vis_square(feat)
```

Filters learned end-to-end from random intialization on the M0 / M1 task:

```python
# the parameters are a list of [weights, biases]
filters = net_end2end.params['conv1'][0].data
vis_square(filters.transpose(0, 2, 3, 1))

# The parameters are fixed once training is finished
```

And the corresponding activations:

```python
# Using end-to-end trained model 
feat = net_end2end.blobs['conv1'].data[0, :]
vis_square(feat)
```

GoogLeNet filters (Convolution 1 7x7 kernel stride 2) learned end-to-end on a non-multiplied dataset , for the M0 / M1 task:

```python
# the parameters are a list of [weights, biases]
filters = net_google.params['conv1/7x7_s2'][0].data
vis_square(filters.transpose(0, 2, 3, 1))

# The parameters are fixed once training is finished
```

And the activations:

```python
# Using GoogLeNet model 
feat = net_google.blobs['conv1/7x7_s2'].data[0, :]
vis_square(feat)
```
